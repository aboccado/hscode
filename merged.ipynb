```python
def check_not_fau1_all(group):
    return (group['관세율구분'] != 'FAU1').all()

not_fau1_all = mergedHsTariff.groupby('품목번호').apply(check_not_fau1_all)
not_fau1_all = not_fau1_all[not_fau1_all]

not_fau1_all_items = not_fau1_all.index.tolist()
```

```python
def check_fau1_any(group):
    return (group['관세율구분'] == 'FAU1').any()

fau1_at_least_once = mergedHsTariff.groupby('품목번호').apply(check_fau1_any)
fau1_at_least_once = fau1_at_least_once[fau1_at_least_once]

fau1_at_least_once_items = set(fau1_at_least_once.index.tolist())

not_fau1_all_items_set = set(not_fau1_all_items)

result = not_fau1_all_items_set - fau1_at_least_once_items
len(result)
```
